### Introduction to Function

Definition:
A function is block of code that performs a specific task. Function help in organizing code, reusing code, and improving readability.

```python
# Syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
# Why functions?
num=24
if num%2 == 0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
def even_odd(num):
    """This function finds even or odd"""
    if num%2 == 0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
# Call the function
even_odd(24)
```

```python
# Function with multiple parameters

def add(a,b):
    return a+b

result = add(2,4)
print(result)
```

```python
# Default Parameters -> if the parameter is not given when calling the function then it uses the default parameter

def greet(name="Guest"):
    print(f"Hello {name} Welcome to the paradise")

greet()
greet("Shyam")
```

```python
# Variable Length Arguments
# Positional and Keywords Arguments

def print_numbers(*shyam):
    for number in shyam:
        print(number)
```

```python
print_numbers(1,2,3,4,"Shyam")
```

```python
# Positional Arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,"Shyam")
```

```python
# Keywords arguments -> all the parameters will be in the form of key value pairs

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"Key:{key}, Value:{value}")
```

```python
print_details(name="Shyam", age="22", country="India")
```

```python
def print_details(*args,**kwargs):
    
    for val in args:
        print(f"Positional argument: {val}")
    
    for key,value in kwargs.items():
        print(f"Key:{key}, Value:{value}")
```

```python
print_details(1,2,3,4,5,"Shyam",name="Shyam", age="22", country="India")
```

```python
# Return statements
def multiply(a,b):
    
    return a*b

multiply(2,3)
```

```python
# Return multiple parameters
def multiply(a,b):
    
    return a*b,a

multiply(2,3)
```
