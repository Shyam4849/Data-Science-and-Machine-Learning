```python
# strings

str = "Shyam Kumar"

for i in str:
    print(i)
```

```python
# While loop
# The while loop continues to run as long as the condition is True.

count = 0

while count<5:
    print(count)
    count = count + 1
```

```python
# Loop control statements

# break -> The break statement exits the loop prematurely

# break statement

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
# continue -> the continue statement skips the current iteration and continues with the next

for i in range(10):
    if i%2 ==0:
        continue
    print(i)
```

```python
# pass -> the pass statement is a null operation; it does nothing

for i in range(5):
    if i==3:
        print("the number is ", i)
        pass
    print(i)
```

```python
# Nested Loop
# a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
# Examples - Calculate the sum of first N natural number using a while and for loop

# while loop

n = 10
sum = 0
count = 1

while(count <= n):
    sum = sum+count
    count += 1
print("Sum of first N natural number: ",sum)
```

```python
# for loop

sum = 0
for i in range(11):
    sum += i
print("Sum of first N natural number: ",sum)
```

```python
# Example - Prime numbers between 1 and 100

for i in range(1,101):
    if i > 1:
        for j in range(2,i):
            if i%2 == 0:
                break
        else:
            print(i)
```

### Conclusion:

Loops are powerful constructs in Python that allows you to execute a block of code multiple times.
